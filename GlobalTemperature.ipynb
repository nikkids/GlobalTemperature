# **ML (K Nearest Neighbor)**

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

Kolom_Kategori = ['Measurement_Region','Klasifikasi']

for kolom in Kolom_Kategori:
  YearlyData[kolom] = label_encoder.fit_transform(YearlyData[kolom])

YearlyData.head(5)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
#Data split
scaler = MinMaxScaler()

#normalisasi
numerik = YearlyData.drop(columns=['Klasifikasi']).select_dtypes(include=['int64','float64']).columns
YearlyData[numerik] = scaler.fit_transform(YearlyData[numerik])

#Bikin X sama Y
X = YearlyData.drop(columns=['Klasifikasi'])
Y = YearlyData['Klasifikasi']

#Data latih sama uji
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)

print(f'X train:{X_train.shape}, Y train:{Y_train.shape}')
print(f'X test:{X_test.shape}, Y test: {Y_test.shape}')
```

```python
YearlyData.head()
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 7, algorithm='auto', leaf_size=30,p=1).fit(X_train,Y_train)
```

```python
from sklearn.metrics import confusion_matrix
y_pred_knn = knn.predict(X_test)
cm = confusion_matrix(Y_test,y_pred_knn)

labels = ['Rendah', 'Mid', 'Tinggi']
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, cmap='Greens', cbar=True, xticklabels=labels, yticklabels=labels)
plt.title('Confusion Matrix')
plt.xlabel('Data Prediksi')
plt.ylabel('Data Sebenarnya')
plt.show()
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
print(f"Precision: {precision_score(Y_test, y_pred_knn,average='macro'):.2f}")
print(f"Recall: {recall_score(Y_test, y_pred_knn,average='macro'):.2f}")
print(f"F1-Score: {f1_score(Y_test, y_pred_knn,average='macro'):.2f}")
```

```python
print(f'Akurasi: {round(accuracy_score(Y_test,y_pred_knn) * 100,1)} %')
```

Menentukan bobot overall temperatur pada suatu tahun apakah rendah, mid, atau tinggi. Matriks korelasi dari klasifikasi data temperature yang memiliki akurasi sebesar 92.9% model klasifikasi memiliki performa yang sangat baik dalam mendeteksi serta memprediksi kelas positif.

# **Periksa Stasioner data dengan uji ADF**

```python
from statsmodels.tsa.stattools import adfuller
X = df['Rata-rata'].values
result=adfuller(X)
print('ADF Statistics: %f'%result[0])
print('p-value: %f'%result[1])
print('Critical Values:')
for key, value in result[4].items():
  print ('\t%s: %.3f'%(key,value))


significance_level = 0.05
if result[1] < significance_level:
    print("Data stasioner")
else:
    print("Data tidak stasioner")
```

Dari hasil uji ADF data rata-rata temperatur tahunan pada daerah Global Land & Ocean menunjukkan bahwa p-value lebih besar dari 0.05 dan nilai ADF lebih besar dari nilai kritis pada taraf signifikan 1% dan 5% sehingga tidak dapat menolak atau menerima H0, maka kesimpulan yang dapat diambil adalah data tidak stasioner.

# **Plot ACF dan PACF**

Plotting Autocorrelation Function (ACF) dan Partial Autocorrelation Function (PACF) setelah data stasioner bertujuan untuk memahami sifat hubungan antar-lag dalam data yang dapat menentukan orde terbaik untuk model.

```python
from statsmodels.tsa.stattools import plot_acf
from statsmodels.tsa.stattools import plot_pacf
acfplot = plot_acf(hasilGlobal_manual, lags=100)
pacfplot = plot_pacf(hasilGlobal_manual, lags=100)
```

Dari hasil plottingan di atas menunjukkan bahwa data stasioner setelah melakukan differencing dengan d = 1 dan uji ADF. Hasil visualisasi plot Autocorrelation Function (ACF) mengidentifikasi komponen Moving Average (MA) dalam model ARMA atau ARIMA. Sedangkan, plot Partial Autocorrelation Function (PACF)mengidentifikasi komponen Autoregressive (AR) dalam model ARMA atau ARIMA. Setelah stasioner, ACF dan PACF akan menunjukkan cutoff setelah beberapa lag, maka jumlah lag signifikan dan dapat digunakan untuk menentukan orde MA (q) dan AR (p).

# **Identifikasi Orde Berdasarkan ACF dan PACF**

```python
from statsmodels.tsa.stattools import acf
acf_values, confint = acf(hasilGlobal_manual, nlags=100, alpha=0.05)

cutoff_lags = []
for lag, (lower, upper) in enumerate(confint[1:], start=1):
    if lower < 0 < upper:
        break
    cutoff_lags.append(lag)

print(f"ACF berhenti signifikan setelah lag ke-{len(cutoff_lags)}")
print(f"Maka nilai q = {len(cutoff_lags)}")
```

Penentuan orde ( q ) (jumlah lag signifikan untuk komponen Moving Average) dilakukan dengan mengevaluasi hasil Autocorrelation Function (ACF). Setiap lag mulai dari lag ke-1 dibandingkan dengan interval kepercayaannya (lower dan upper). Semua lag yang memiliki korelasi di luar interval kepercayaan dianggap signifikan kemudian disimpan hingga ditemukan lag pertama yang tidak signifikan. Berdasarkan analisis ini, diperoleh nilai ( q = 34 ), yang berarti lag signifikan untuk komponen Moving Average (MA) berhenti setelah lag ke-34, atau terjadi cut-off pada lag tersebut.

```python
from statsmodels.tsa.stattools import pacf
acf_values, confint = pacf(hasilGlobal_manual, nlags=100, alpha=0.05, method = 'yw')

cutoff_lags = []
for lag, (lower, upper) in enumerate(confint[1:], start=1):
    if lower < 0 < upper:
        break
    cutoff_lags.append(lag)

print(f"PACF berhenti signifikan setelah lag ke-{len(cutoff_lags)}")
print(f"Maka nilai p = {len(cutoff_lags)}")
```

Penentuan orde ( p ) (jumlah lag signifikan untuk komponen Autoregressive (AR)) dilakukan dengan mengevaluasi hasil Partial Autocorrelation Function (PACF). Setiap lag mulai dari lag ke-1 dibandingkan dengan interval kepercayaannya (lower dan upper). Semua lag yang memiliki korelasi di luar interval kepercayaan dianggap signifikan kemudian disimpan hingga ditemukan lag pertama yang tidak signifikan. Berdasarkan analisis ini, diperoleh nilai ( p = 7 ), yang berarti lag signifikan untuk komponen Partial Autocorrelation Function (PACF) berhenti setelah lag ke-7, atau terjadi cut-off pada lag tersebut.

```python
acf_values, confint = acf(hasilGlobal_manual, nlags=100, alpha=0.05)

cutoff_lags = []
for lag, (lower, upper) in enumerate(confint[1:], start=1):
    if upper < 0 or 0 < lower:
      cutoff_lags.append(lag)

print(f"{cutoff_lags}")
```
